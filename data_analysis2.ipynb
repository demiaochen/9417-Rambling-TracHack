```python
import json
import seaborn as sns
import matplotlib as plt
import pandas as pd

# Opening JSON file
with open('manf_probability.json') as json_file1:
    data1 = json.load(json_file1)
 
    # Print the type of data variable
    print("Type:", type(data1))
 
    # Print the data of dictionary
    print(data1)
```

```python
# Opening JSON file
with open('state_probability.json') as json_file2:
    data2 = json.load(json_file2)
 
    # Print the type of data variable
    print("Type:", type(data2))
        # Print the data of dictionary
    print(data2)
```

```python
# Opening JSON file
with open('os_probability.json') as json_file3:
    data3 = json.load(json_file3)
 
    # Print the type of data variable
    print("Type:", type(data3))
#         Print the data of dictionary
    print(data3)
```

```python
df_m = pd.DataFrame(data1.items(), columns=['Manufacturer', 'Values'])
df_s = pd.DataFrame(data2.items(), columns=['State', 'Values'])
df_os = pd.DataFrame(data3.items(), columns=['OS', 'Values'])
display(df_m)
display(df_s)
display(df_os)
```

```python
sns.set_color_codes("pastel")
p1 = sns.barplot(x='OS', y="Values", data=df_os,
            label="Total", color="b")
p1.set(xticklabels=[])
p1.set(ylabel='Probability')
p1.set(xlabel='Operating System')
plt.pyplot.savefig('os_dist_full.png', 
           transparent=True)
```

```python
sns.set_color_codes("pastel")
p1 = sns.barplot(x='Manufacturer', y="Values", data=df_m,
            label="Total", color="b")
p1.set(xticklabels=[])
p1.set(ylabel='Probability')
p1.set(xlabel='Manufacturer')
plt.pyplot.savefig('manf_dist_full.png', 
           transparent=True)
```

```python
sns.set_color_codes("pastel")
p1 = sns.barplot(x='State', y="Values", data=df_s,
            label="Total", color="b")
p1.set(xticklabels=[])
p1.set(ylabel='Probability')
p1.set(xlabel='State')
plt.pyplot.savefig('state_dist_full.png', 
           transparent=True)
```
