# Process categorical data for Manufacturer, State, OS, language and opt_out_sms
### Replace null values with string 'NONE'

```python
def standardiseFormat(df, col):
    out = df.copy()
    # print(out[col].isnull().sum())
    out[col] = out[col].fillna('<NO_ENTRY>')
    out[col] = out[col].apply(lambda x: str(x).lower())
    # print(out[col].isnull().sum())
    return out
```

```python
cols = ['manufacturer', 'state', 'operating_system', 'language_preference']

df2 = standardiseFormat(df2, 'manufacturer')
df2 = standardiseFormat(df2, 'state')
df2 = standardiseFormat(df2, 'operating_system')
df2 = standardiseFormat(df2, 'language_preference')

display(df2)
```

# Use Target Encoding to encode categorical data, and merge new columns to dataframe

```python
def TargetEncoding(df, col):
    m_encoder = ce.TargetEncoder()
    out = m_encoder.fit_transform(df[col], df['ebb_eligible'])
    new_col = col + '_enc'
    out.columns = [new_col]
    return out
```

```python
df2_m = TargetEncoding(df2, 'manufacturer')
df2_s = TargetEncoding(df2, 'state')
df2_os = TargetEncoding(df2, 'operating_system')
df2_l = TargetEncoding(df2, 'language_preference')

# Now add to dataframe
df2 = pd.concat([df2, df2_m], axis=1)
df2 = pd.concat([df2, df2_s], axis=1)
df2 = pd.concat([df2, df2_os], axis=1)
df2 = pd.concat([df2, df2_l], axis=1)
display(df2)
```

# Map the encoded value for each unique category in the dictionary
#### *  Requires the dataframe to contain the encoded column

```python
# map the encoded values to the dictionary of unique keys
# Requires the dataframe to already contain the encoded column
def fillCatValues(df, col, cat_dict):
    col_enc = col + '_enc'
    for index, row in df.iterrows():
        key = row[col]
        if key in cat_dict and cat_dict[key] is None:
            cat_dict[key] = row[col_enc]
```

```python
# fill dict values
fillCatValues(df2, 'manufacturer', cat_m)
fillCatValues(df2, 'state', cat_s)
fillCatValues(df2, 'operating_system', cat_os)
fillCatValues(df2, 'language_preference', cat_l)
```

### Process dates columns. Assumes that all dates are valid

```python
'''
    Split dates into their respective year, month and day
    Creates 3 new columns for each date column (last_redemption_date and first_activation_date)
    Drops original column
'''
def processDates(df_in):
    df = df_in.copy()

    # Convert to datetime format
    df['last_redemption_date'] = pd.to_datetime(df['last_redemption_date'], format = '%Y-%m-%d', errors = 'coerce')
    # Break up year, month, day
    df['last_redemption_year'] = df['last_redemption_date'].dt.year
    df['last_redemption_month'] = df['last_redemption_date'].dt.month
    df['last_redemption_day'] = df['last_redemption_date'].dt.day
    # Drop original column
    df = df.drop(['last_redemption_date'],axis=1)

    # Convert to datetime format
    df['first_activation_date'] = pd.to_datetime(df['first_activation_date'], format = '%Y-%m-%d', errors = 'coerce')
    # Break up year, month, day
    df['first_activation_year'] = df['first_activation_date'].dt.year
    df['first_activation_month'] = df['first_activation_date'].dt.month
    df['first_activation_day'] = df['first_activation_date'].dt.day
    # Drop original column
    df = df.drop(['first_activation_date'],axis=1)
    
    return df
# print(df.columns)
```

```python
df2 = processDates(df2)
eval2 = processDates(eval2)
display(df2)
display(eval2)
```

### Process Opt-out columns
{opt_out_email,  opt_out_loyalty_email, opt_out_loyalty_sms, opt_out_mobiles_ads, opt_out_phone}

opt_out_mobiles_ads contains two values, 1 and -1. \
Perform One-Hot Encoding on this column to extract -1 values into a new column

```python
# create new column where df2['opt_out_mobiles_ads'] == -1
# replace -1 values in df2['opt_out_mobiles_ads'] with 0
def encodeOptMobileAds(df_in):
    df = df_in.copy()
    df['opt_out_mobile_ads_neg'] = np.where(df['opt_out_mobiles_ads'] == -1, 1, 0) 
    # display(df[['opt_out_mobiles_ads', 'opt_out_mobile_ads_neg']])
    df['opt_out_mobiles_ads'] = df['opt_out_mobiles_ads'].map({1: 1, -1: 0})
    # display(df[['opt_out_mobiles_ads', 'opt_out_mobile_ads_neg']])
    return df
```

```python
def processOpt(df_in):
    df = df_in.copy()
    opt_out = ['opt_out_email', 'opt_out_loyalty_email', 'opt_out_loyalty_sms', 'opt_out_mobiles_ads', 'opt_out_phone']
    # fill NaN columns with 0
    for option in opt_out:
        df[option].fillna(0, inplace=True)
    return df
```
