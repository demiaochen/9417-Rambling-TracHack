```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

df_ebb1 = pd.read_csv("../data/ebb_set1.csv")
df_ebb2 = pd.read_csv("../data/ebb_set2.csv")
df = pd.concat([df_ebb1, df_ebb2], ignore_index=True)

dfp = pd.read_csv("js_plotsets.csv")
# display(df)
display(dfp)
```

```python
plt.figure(figsize=(15,10))
ax = sns.relplot(x="total_redemptions", y="ebb_eligible", data = dfp, color = 'red',sizes=(100, 200), height=7, marker='o')
ax.set_axis_labels("redemptions", "ebb_eligible")
```

```python
dfp.columns
```

```python
features = ['opt_out_email', 'opt_out_loyalty_email', 'opt_out_loyalty_sms',
           'opt_out_mobiles_ads', 'opt_out_phone', 'opt_out_mobile_ads_neg']


for feature in features:
    snsplot = sns.catplot(x=feature,hue='ebb_eligible',data=dfp,kind="count")
    fname = "plots/"+ feature + "_ebb.png"
    snsplot.savefig(fname)
```

```python
features = ['last_redemption_year',
       'last_redemption_month', 'last_redemption_day', 'first_activation_year',
       'first_activation_month', 'first_activation_day']

for feature in features:
    snsplot = sns.catplot(x=feature,hue='ebb_eligible',data=dfp,kind="count")
    plt.xticks(rotation=90)
    fname = "plots/"+ feature + "_ebb.png"
    snsplot.savefig(fname)
```

```python
sns.catplot(x="ebb_eligible", y="tenure", data=dfp)
fname = "plots/tenure_ebb.png"
snsplot.savefig(fname)
```
